## Get the data

```python
import pandas as pd
poll_url = 'http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv'
polls = pd.read_csv(poll_url)
polls
```

```python
polls.info()
```

```python
polls.columns
```

## Sort the data

```python
polls.sort_values('enddate', ascending=False, inplace=True)
polls.head()
```

## How to continue statements

### With implicit continuation

```python
polls.sort_values(['state','startdate'], 
                  ascending=False,
                  inplace=True)
polls.head()
```

### With explicit continuation

```python
polls.sort_values(['state','startdate'], \
                  ascending=False, \
                  inplace=True)
polls.head()
```
