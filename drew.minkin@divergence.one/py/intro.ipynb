```python
#;ist of words ( string)
kids =['John', 'Sandy', 'Pat']
#control flow - loop - while
tgt = "t"
ctr = 0
while (ctr < len(kids)): 
    if tgt in kids[ctr]:
        ctr = ctr-1
    ctr = ctr+1    
    print("{0}: {1}".format(kids[ctr],ctr))
```
