```python
from bs4 import BeautifulSoup
import requests
import re
from pprint import pprint
import pandas as pd
import numpy as np
import urllib.request 
```

```python
scriptdf=pd.read_csv('/content/drive/MyDrive/DSI_capstone/scriptdf',index_col='Title')
```

```python
scriptdf.drop(columns='Unnamed: 0',inplace =True)
```

```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

worksheet = gc.open('movies').sheet1

rows = worksheet.get_all_values()

movies = pd.DataFrame.from_records(rows,)
```

```python
movies.columns = movies.iloc[0]
```

```python
movies=movies.drop(0)
```

```python
movies.describe()
```

```python
scriptdf
```

```python
result = pd.merge(movies, scriptdf, how="left", on=["Title"])
```

```python
result.describe()
```

```python
result.dropna(inplace=True)
```

```python
movielinks=result.sort_values('Rating',ascending=False)
```

```python
movielinks.link
```

```python
movielinks['script']=np.nan
```

```python
movielinks.head()
```

```python
for x in movielinks.link:
  url = x
  page = requests.get(url)
  soup = BeautifulSoup(page.content, "html.parser")
  results=soup.get_text()
  movielinks.script= results
```

```python
movielinks.to_csv('/content/drive/MyDrive/DSI_capstone/scripttext')
```
