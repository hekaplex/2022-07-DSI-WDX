```sql
--How many items with ListPrice more than $1000 have been sold?
SELECT COUNT(*)
FROM Sales.SalesOrderDetail 
WHERE  UnitPrice > 1000
```

```sql
--Give the Store Name of those customers with Total orders over $100000. Include the subtotal plus tax plus freight.
SELECT Store.Name, Sales.SalesOrderHeader. TotalDue
FROM Sales.Store Store
JOIN Sales.SalesOrderHeader
ON Sales.SalesOrderHeader.SalesPersonID = Store.SalesPersonID
WHERE TotalDue > 100000
ORDER BY Store.Name
```

```sql
--A "Single Item Order" is a customer order where only one item is ordered. Show the SalesOrderID and the UnitPrice for every Single Item Order.
SELECT SalesOrderID, UnitPrice
FROM Sales.SalesOrderDetail
WHERE OrderQty <= 1
```
