```sql

/*
Adventureworks2019 SQL Exercises
Adapted from SQLZoo

Easy SQL Questions
How many/// items with ListPrice/// more than $1000 have been sold////?///////
Give the Store Name//// of those customers with Total orders over $100000////. Include the subtotal plus tax plus freight////.

Medium SQL Questions
A "Single Item Order" is a customer order where only one item is ordered. Show the SalesOrderID and the UnitPrice for every Single Item Order.
Where did the racing socks go? List the product name and the Store Name for all Customers who ordered ProductModel 'Racing Socks'.
Show the product description for culture 'fr' for product with ProductID 736.
Use the SubTotal value in SaleOrderHeader to list orders from the largest to the smallest. For each order show the Store Name and the SubTotal and the total weight of the order.
How many products in ProductCategory 'Components' have been sold to an address in 'London'?

Hard SQL Questions
For every customer with a 'Main Office' in Dallas show AddressLine1 of the 'Main Office' and AddressLine1 of the 'Shipping' address - if there is no shipping address leave it blank. Use one row per customer.
For each order show the SalesOrderID and SubTotal calculated three ways: A) From the SalesOrderHeader B) Sum of OrderQtyUnitPrice C) Sum of OrderQtyListPrice
Show the best selling item by value.
Show how many orders are in the following ranges (in $):

RANGE	Num Orders
0 - 99	
100- 999	
1000-9999	
10000+	
List the name and addresses of companies containing the word 'Bike' (upper or lower case) and companies containing 'cycle' (upper or lower case). Ensure that the 'bike's are listed before the 'cycles's.
Show the total order value for each CountryRegion. List by value with the highest first.
Find the best customer in each region.
*/


--Easy questions


SELECT 
sum(pth.Quantity) AS TOTAL_SOLD
FROM 
    Production.ProductListPriceHistory PPLH
    join Production.TransactionHistory PTH
        ON PPLH.ProductID = PTH.ProductID
    WHERE   ListPrice > 1000
    --Group by pth.Quantity




SELECT Distinct
ss.name
--,ssoh.TotalDue

From 
Sales.SalesOrderHeader SSOH
left join Sales.Customer SC
    on SSOH.CustomerID = SC.CustomerID
right Join Sales.Store SS
    ON SC.StoreID = SS.BusinessEntityID
    Where SSOH.TotalDue > 100000
    --order by SSOH.TotalDue Desc
```
